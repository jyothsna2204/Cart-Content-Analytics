```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
df = pd.read_csv('Market_Basket_Optimisation.csv',header=None)
df.head()
```

```python
df.fillna(0,inplace=True)
df.head()
```

```python
transactions = []
for i in range(0,len(df)):
    transactions.append([str(df.values[i,j]) for j in range(0,20) if str(df.values[i,j])!='0'])
```

```python
transactions[0]
```

```python
rules = apriori(transactions,min_support=0.003,min_confidance=0.2,min_lift=3,min_length=2)
rules
```

```python
# all rules need to be converted in a list..
Results = list(rules)
Results
```

```python
df_results = pd.DataFrame(Results)
df_results
```

```python
support = df_results.support
```

```python
first_values = []
second_values = []
third_values = []
fourth_value = []
for i in range(df_results.shape[0]):
    single_list = df_results['ordered_statistics'][i][0]
    first_values.append(list(single_list[0]))
    second_values.append(list(single_list[1]))
    third_values.append(single_list[2])
    fourth_value.append(single_list[3])
```

```python
lhs = pd.DataFrame(first_values)
rhs= pd.DataFrame(second_values)
confidance=pd.DataFrame(third_values,columns=['Confidance'])
lift=pd.DataFrame(fourth_value,columns=['lift'])
#concat all list together in a single dataframe
df_final = pd.concat([lhs,rhs,support,confidance,lift], axis=1)
df_final
```

```python
df_final.columns = ['lhs',1,2,3,4,'support','confidance','lift']
df_final
```

```python
df_final['concatenated'] = df_final[[1, 2, 3, 4]].apply(lambda x: ', '.join(x.dropna().astype(str)), axis=1)
# Append the concatenated column without null values
df_final['rhs'] = df_final['concatenated'].apply(lambda x: x if x else None)
df_final
```

```python
df_final.drop(columns=[1,2,3,4,'concatenated'],inplace=True)
```

```python
df_final.drop(columns=["support","confidance","lift"],inplace=True)
```

```python
lhs_rhs_dict = {}

# Iterate over DataFrame rows
for index, row in df_final.iterrows():
    lhs = row['lhs']
    rhs = row['rhs']
    
    # If lhs already exists in dictionary, append rhs value to existing list
    if lhs in lhs_rhs_dict:
        lhs_rhs_dict[lhs].append(rhs)
    # If lhs is new, create a new list with rhs value
    else:
        lhs_rhs_dict[lhs] = [rhs]

# Convert dictionary to DataFrame
df_final = pd.DataFrame(lhs_rhs_dict.items(), columns=['lhs', 'rhs'])

print(df_final)
```
